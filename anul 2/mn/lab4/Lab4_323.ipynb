```python
# Nume student:
# CMMP - Problema celor mai mici patrate
# TORT - triangularizare ortogonala cu reflectori 
import numpy as np
rng = np.random.default_rng(0)

# ----------- Input ---------------------------
#  !! Sectiunea Input nu se modifica !!
m = 7 # nr de linii
n = 4 # nr de coloane
A = rng.integers(1,11, size=(m,n))
A = np.triu(A,-1)
A[m-1,:] = rng.integers(1,10, size=(1,n))
A = A.astype(float)
b = rng.integers(10, size=(m,1)).astype(float)
#-----------------------------------------------
print('------------A-------------\n', A)

sol = np.linalg.pinv(np.copy(A))@np.copy(b)

# Descompunerea QR comanda in Python
q,r = np.linalg.qr(np.copy(A))
print('--------q-----\n', q)
print('--------r-----\n',r)

def  TORT (A):
    m,n  = np.shape(A)
    U = np.zeros((m,n))
    p = min(m-1,n)
    beta = np.zeros((p,1))
    for k in range(p):
        # Calcularea lui sigma
        s = A[k+1][k]**2 +A[m-1][k]**2 + A[k][k]**2 #nu mai avem nevoie de for pentru suma
        sigma = np.sign(A[k,k])*np.sqrt(s)
        if sigma != 0:
            # Calcularea vectorilor householder
            U[k][k] = A[k][k] +sigma
            U[k+1][k] = A[k+1][k]#restul elementelor sunt nule
            U[m-1][k] = A[m-1][k]          
            beta[k] = sigma*U[k,k]
            #Aplicarea reflectorului Householder pe A
            for j in range(k,n):
                tau = U[k][k]*A[k][j] + U[k+1][k]*A[k+1][j] + U[m-1][k]*A[m-1][j]
                tau =tau/beta[k]
                A[k,j]=A[k,j]-tau*U[k,k]
                A[k+1,j]=A[k+1,j]-tau*U[k+1,k]
                A[m-1,j]=A[m-1,j]-tau*U[m-1,k]
                    
    return U, A, beta

# Algoritmul UTRIS
def Utris(U, b):
    n = len(U)
    x = np.zeros((n,1))
    for i in range(n-1,-1, -1):
        s = b[i]
        for j in range(i+1,n):
            s = s -U[i][j]*x[j]
        x[i] = s/U[i][i]
    return x


#Triangularizarea ortogonala a lui A
U,R,beta  = TORT(A)
print('----------U------------\n', U)
print('-----------R-----------\n',R)

# Aplicarea Reflectorilor asupra lui b
for k in range(n):
    tau = U[k][k]*b[k] + U[k+1][k]*b[k+1]+ U[m-1][k]*b[m-1]
    tau = tau/beta[k]
    b[k]= b[k]- tau*U[k][k]
    b[m-1] = b[m-1] - tau*U[m-1][k]
    b[k+1] = b[k+1] - tau*U[k+1][k]

#Calcularea solutiei CMMP
x  =Utris(R[0:n,:], b[0:n,0])

print('-----------x------------\n',x)
print('----------sol------------\n', sol)
```
